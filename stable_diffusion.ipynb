```python
from diffusers import StableDiffusionPipeline
import torch


def generate_image(prompt: str, output_path: str = "generated_image.png"):

    print("Loading model...")
    pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1-base", torch_dtype=torch.float16)
    pipe.to("cuda")
    print("Generating image...")
    image = pipe(prompt).images[0]

    image.save(output_path)
    print(f"Image saved at {output_path}")

if __name__ == "__main__":
    text_prompt = "a futuristic city with flying cars, neon lights, and skyscrapers"
    generate_image(text_prompt, "futuristic_city.png")
```
